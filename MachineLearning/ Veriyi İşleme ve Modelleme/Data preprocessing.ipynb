```python

# 'Ekran Boyutu' sütunundaki boş değerlere sahip satırları sil
df = df.dropna(subset=['Ekran Boyutu'])
```

```python
# 'Çözünürlük' sütunundaki veri sayısını bul
cozunurluk_veri_sayisi = df['Çözünürlük'].count()
```

```python
df = df.dropna(subset=['Çözünürlük'])
```

```python
df.dropna(inplace=True)
```

```python

# 'Belirtilmemiş', 'None', 'Yok', 'SSD Yok' gibi geçersiz veya eksik değerleri NaN ile değiştirme fonksiyonu
def replace_invalid_with_nan(value):
    if value in [None, 'None', 'Yok', 'SSD Yok', 'Belirtilmemiş']:
        return np.nan
    try:
        return float(value)
    except ValueError:
        return np.nan
```

```python
#fiyat sütunundaki TL ifadesini silme 
df['price'] = df['price'].str.replace(' TL', '')
```

```python
# 'Ram (Sistem Belleği)' ve 'SSD Kapasitesi' sütunlarını temizle
df['Ram (Sistem Belleği)'] = df['Ram (Sistem Belleği)'].str.replace(' GB', '').apply(replace_invalid_with_nan)
df['SSD Kapasitesi'] = df['SSD Kapasitesi'].replace(['SSD Yok', 'Yok', 'Belirtilmemiş', 'None'], np.nan)
df['SSD Kapasitesi'] = df['SSD Kapasitesi'].str.replace(' GB', '').str.replace(' TB', '000').str.replace(' ', '').apply(replace_invalid_with_nan)

```

```python
# 'Ekran Boyutu' sütununu temizle
df['Ekran Boyutu'] = df['Ekran Boyutu'].str.replace(' inç', '').str.replace(',', '.').apply(replace_invalid_with_nan)
```

```python
# 'Çözünürlük' sütununu temizle ve iki ayrı sayısal sütuna böl
resolution = df['Çözünürlük'].str.split('x', expand=True)
df['Çözünürlük Genişlik'] = resolution[0].str.strip().apply(replace_invalid_with_nan)
df['Çözünürlük Yükseklik'] = resolution[1].str.strip().apply(replace_invalid_with_nan)
df.drop('Çözünürlük', axis=1, inplace=True)
```

```python
# Kategorik değerleri standartlaştırma ve kodlama fonksiyonu
def standardize_and_encode(column, mapping):
    # Sütundaki değerleri küçük harfe çevir ve boşlukları kaldır
    standardized = column.str.lower().str.replace(' ', '')
    # Eşleşen değerler için kodları döndür, eşleşmeyenler için 0
    return standardized.map(mapping).fillna(0)
```
